```python
columns_to_drop = [
    'belongs_to_collection',
    'budget',
    'homepage',
    'imdb_id',
    'overview',
    'poster_path',
    'production_countries',
    'revenue',
    'runtime',
    'spoken_languages',
    'status',
    'tagline',
    'video',
    'original_language',
]
```

```python
df = metadata.drop(columns_to_drop, axis=1)
```

```python
df
```

```python
df['genres'] = df['genres'].apply(ast.literal_eval)
```

```python
df = df[df['genres'].apply(len) > 0]  #filter out movies with no genres defined
```

```python
df
```

```python
all_genres = {}
for genres in df['genres']:
    for genre in genres:
        if genre['name'] not in all_genres:
            all_genres[genre['name']] = {'id': genre['id'], 'count': 1}
        else:
            all_genres[genre['name']]['count'] += 1
all_genres
```

```python
%matplotlib widget
import seaborn as sns
genres = {}
for genre in all_genres:
    genres[genre] = all_genres[genre]['count']

X = [key for key in genres.keys()]
Y = [val for val in genres.values()]
sns.barplot(x = X, y = Y)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
def isNaT(series, is_not = False):
    ret_lis = []
    for date in series:
        if type(date) == pd._libs.tslibs.timestamps.Timestamp:
            if is_not:
                ret_lis.append(False)
            else:
                ret_lis.append(True)
        else:
            if is_not:
                ret_lis.append(True)
            else:
                ret_lis.append(False)
    return np.array(ret_lis)
```

```python
df[isNaT(df['release_date'])].isnull().sum()
```

```python
df = df[isNaT(df['release_date'])]
```
