```python
def get_genres_list(row):
    genres_list = []
    for genre_dict in row:
        genres_list.append(genre_dict['name'])
    return genres_list
```

```python
metadata['genres'] = metadata['genres'].apply(get_genres_list)
```

```python
def get_production_companies(row):
    production_list = []
    for production_dict in row:
        production_list.append(production_dict['name'])
    return production_list
```

```python
metadata['production_companies'][0]
```

```python
metadata['production_companies'] = metadata['production_companies'].apply(get_production_companies)
```

```python
def is_not_null_list(row):
    return row != []
```

```python
metadata = metadata[metadata['production_companies'].apply(is_not_null_list)]
```

```python
def tokenize_overview(row):
    return row.split(' ')
```

```python
metadata['overview'] = metadata['overview'].apply(tokenize_overview)
```

```python
def get_n_top_cast(n):
    def get_top_cast(row):
        top_cast = []
        i = 0
        for cast in row[0]:
            i += 1
            top_cast.append(cast['name'].replace(' ', ''))
            if i >= n:
                break
        return top_cast
    return get_top_cast
```

```python
metadata['cast'] = metadata['credits'].apply(get_n_top_cast(5))
```

```python
def get_director(row):
    directors = []
    for job in row[1]:
        if job['job'] == 'Director':
            directors.append(job['name'].replace(' ', ''))
    return directors
```

```python
metadata['directors'] = metadata['credits'].apply(get_director)
```

```python
metadata['tags'] = metadata['genres'] + metadata['overview'] + metadata['cast'] + metadata['directors'] + metadata['production_companies']
```

```python
final_data = metadata.drop(['genres', 'overview', 'production_companies', 'title', 'credits', 'cast', 'directors'], axis=1)
```

```python
final_data['tags'] = final_data['tags'].apply(lambda x: ' '.join(x))
```

```python
final_data.to_csv('dataset/processed_data.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(final_data['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# similarities = cosine_similarity(vectors)
chunk_size = 500 
matrix_len = vectors.shape[0] # Not sparse numpy.ndarray

def similarity_cosine_by_chunk(start, end):
    if end > matrix_len:
        end = matrix_len
    return cosine_similarity(X=vectors[start:end], Y=vectors) # scikit-learn function

chunks = []

for chunk_start in range(0, matrix_len, chunk_size):
    cosine_similarity_chunk = similarity_cosine_by_chunk(chunk_start, chunk_start+chunk_size)
    chunks.append(cosine_similarity_chunk)    
```
